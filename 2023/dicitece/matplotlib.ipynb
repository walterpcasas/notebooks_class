```python
plt.plot(x[:80], x[:80]*79, 'g-')
plt.plot(x[79:], x[79:]**2, 'r--')
plt.show()
```

```python
plt.figure(figsize=(4, 8))
plt.plot(x[:80], x[:80]*79, 'g-')
plt.plot(x[79:], x[79:]**2, 'r--')
plt.show()
```

```python
plt.figure(figsize=(4, 8))
plt.title('Ingresos de la companhia')
plt.plot(x[:80], x[:80]*79, 'g-')
plt.plot(x[79:], x[79:]**2, 'r--')
plt.show()
```

```python
plt.figure(figsize=(4, 8))
plt.title('Ingresos de la companhia')
plt.plot(x[:80], x[:80]*79, 'g-')
plt.plot(x[79:], x[79:]**2, 'r--')
plt.xlabel('Nro trabajadores')
plt.ylabel('Ingresos en dolares')
plt.show()
```

```python
plt.figure(figsize=(4, 8))
plt.title('Ingresos de la companhia')
plt.plot(x[:80], x[:80]*79, 'g-', label='Real')
plt.plot(x[79:], x[79:]**2, 'r--', label='Predicho')
plt.xlabel('Nro trabajadores')
plt.ylabel('Ingresos en dolares')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(4, 8))
plt.title('Ingresos de la companhia')
plt.plot(x[:80], x[:80]*79, 'g-', label='Real')
plt.plot(x[79:], x[79:]**2, 'r--', label='Predicho')
plt.xlabel('Nro trabajadores')
plt.ylabel('Ingresos en dolares')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(4, 8))
plt.title('Ingresos de la companhia')
plt.plot(x[:80], x[:80]*79, 'g-', label='Real')
plt.plot(x[79:], x[79:]**2, 'r--', label='Predicho')
plt.xlabel('Nro trabajadores')
plt.ylabel('Ingresos en dolares')
plt.grid()
plt.legend()
plt.savefig('ingresos.png')
plt.show()
```

```python
plt.figure(figsize=(4, 8))
plt.title('Ingresos de la companhia')
plt.plot(x[:80], x[:80]*79, 'g-', label='Real')
plt.plot(x[79:], x[79:]**3, 'r--', label='Predicho')
plt.xlabel('Nro trabajadores')
plt.ylabel('Ingresos en dolares')
plt.grid()
plt.legend()
plt.savefig('ingresos2.png')
plt.show()
```

# Histograma

```python
v = np.random.normal(15, 5, 50)
v
```

```python
plt.hist(v)
plt.show()
```

```python
plt.hist(w)
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.hist(v)

plt.subplot(1, 2, 2)
plt.hist(w)

plt.show()
```

```python
plt.figure(figsize=(13, 4))

plt.subplot(1, 3, 1)
plt.hist(v)

plt.subplot(1, 3, 2)
plt.hist(w)

plt.subplot(1, 3, 3)
plt.scatter(x, y)

plt.show()
```

```python
plt.figure(figsize=(13, 4))

plt.subplot(131)
plt.hist(v)

plt.subplot(132)
plt.hist(w)

plt.subplot(133)
plt.scatter(x, y)

plt.show()
```

# Grafico torta

```python
pct = np.array([20, 15, 55, 10]) 
```

```python
plt.pie(pct)
plt.show()
```

```python
sist_operativos = ['OSX', 'Linux', 'Windows', 'Others']
```

```python
plt.pie(pct, labels=sist_operativos)
plt.show()
```

```python
plt.pie(pct, labels=sist_operativos)
plt.legend()
plt.show()
```

```python
pct = np.array([20, 25, 45, 10]) 
```

```python
plt.pie(pct, labels=sist_operativos, autopct='%1.0f%%')
plt.legend()
plt.show()
```
